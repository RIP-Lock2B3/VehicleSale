```python
from sklearn.model_selection import train_test_split
x=data[['Year','Kilometers_Driven','Engine','Power','Seats','L_Location','L_Fuel','L_Transmission','L_Owner']].values
y=data['Price'].values
```

```python
XTrain,XTest,YTrain,YTest=train_test_split(x,y,test_size=0.8,random_state=0)
```

```python
XTest
```

```python
from sklearn.linear_model import LinearRegression
linearRegressor=LinearRegression()
linearRegressor.fit(XTrain,YTrain)
```

```python
YPrediction=linearRegressor.predict(XTest)
```

```python
from sklearn.metrics import f1_score,accuracy_score
f1_score(YTest,YPrediction,average='macro')
```
